
# Before getting started Import all the packages that are necessary for Visualization

```python
import pandas as pd
import numpy as np  
import seaborn as sb
import matplotlib.pyplot as plt
train=pd.read_csv('train.csv')
```

```python
train.head()
```

```python
# Finding the Percent of Null Values in the Data
(train.isnull().sum()/len(train))*100
```

The above simple code reveals the attribute which has null values and routes the user for the further step

```python
#Visualizing the Survivers with respect to the Pclass
sb.barplot(x='Pclass', y='Survived', data=train)
```

```python
#Now lets a new term "Hue" and lets see what it does to the data
sb.barplot(x='Pclass', y='Survived',hue='Embarked', data=train)
```

In the above we can see that, with the introduction of "Hue" it displayed the segregation of the data with respect to the attribute

```python
# Pie chart descibing the fare
fare=train.groupby(['Embarked'])['Fare'].mean()
plt.pie(fare,labels=fare.index,autopct='%0.1f%%')
```

Make the chart by making an condition and make sure that it would be a meaningful , In the last "%0.1f%%" depicts about rounding of the Percentage in the chart , increase in the value , increase in decimal points

# Jointplot

Jointplot is the paramount in the visualization as it depicts much about the distribution of pattern in the data, and perfectly conveys the inference even to a layman

It has to plotted between any two attributes and the total information about the outliers in the data

```python
sb.jointplot(train['Fare'],train['Age'])
```

Introducing the KDE plot into the visualization

```python
sb.jointplot(train['Fare'],train['Age'],kind='kde')
#It shows the dense region where most of the data lies
```

```python
plt.scatter(train['Fare'],train['Age'])
```

```python
# This shows about the distribution of the data and helps in knowing the Quartile portions of the data
sb.violinplot(train['Age'],inner='quartile')
```

```python
data_num=train._get_numeric_data()
correlation=data_num.corr()
correlation
sb.heatmap(correlation)
```

```python
data_num=train._get_numeric_data()
data_num=data_num.dropna()
sb.pairplot(data_num)
```
